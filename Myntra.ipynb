```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
df
```

# 1. How many unique products are there in the dataset? (5 marks)
You need to find product ID of each product from product_link and then find the number of unique product ids

Finding Unique Products as per their Product ID
Removing Products with Duplicate Product ID

```python
df['product_id'] = df['product_link'].str.extract(r'/(\d+)/buy')
```

```python
df['product_id'].nunique()
```

```python
df = df.drop_duplicates('product_id')
```

# 6. List top 5 brands with maxiumum average rating of products? (2 marks)
Neglect the products which are not reviewed by any person

```python
brand_ratings = df.groupby('brand_name')['rating'].mean()

# sort the brands in descending order of their mean rating and select the top 5 brands
top_5_brands = brand_ratings.sort_values(ascending=False).head(5)

# print the top 5 brands
print(top_5_brands)
```

# 7. Plot a histogram of the distribution of ratings of following brand. (3 marks)
H&M
max
Puma
MANGO
Neglect the products which are not reviewed by any person

```python
brands = ['H&M', 'max', 'Puma', 'MANGO']
for brand in brands:
    brand_df = df[df['brand_name'] == brand]
    brand_df = brand_df[brand_df['rating_count'] > 0]
    plt.hist(brand_df['rating'], bins=10)
    plt.title(f'Histogram of {brand} ratings')
    plt.xlabel('Rating')
    plt.ylabel('Frequency')
    plt.show()
```

# 8. What is the distribution of discounts by brand for specific brands mentioned below? (2 marks)
H&M
max
Puma
MANGO
Neglect the products which are not reviewed by any person. Neglect the products which are not on any discount.

```python
brands = ['H&M', 'max', 'Puma', 'MANGO']
df_filtered = df[df['brand_name'].isin(brands)]

# filter out products that are not reviewed by any person and are not on any discount
df_filtered = df_filtered[(df_filtered['rating'] > 0) & (df_filtered['discount_amount'] > 0)]

# group the products by brand and calculate the distribution of discounts for each brand
discount_distribution = df_filtered.groupby('brand_name')['discount_amount'].value_counts(normalize=True)

# print the distribution of discounts for each brand
discount_distribution
```

# 9. Is there any trend or pattern you can find between discount percent and rating of the product? (3 marks)

```python
df_dist=pd.DataFrame(list(discount_distribution.items()),columns = ['brand_name','discount_amount'])
```

```python
corr = df_dist.corr()

sns.heatmap(corr, annot=True)
```
